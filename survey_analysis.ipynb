```python
data['agea'].hist(bins=30)
plt.xlabel("Age (years)")
plt.ylabel("Frequency")
plt.title("Age distribution")
plt.show()
```

### Distribution of Age

The age distribution shows a broad spread across adult respondents with a clear concentration between approximately the ages of 40 and 70 years. Frequencies are lower amongst younger adults (below ~30) and decline more sharply at higher ages (above ~70).

This indicates that the sample is dominated by middle-aged individuals, with particularly limited representation of older age groups, which may affect the interpretation of age-related patterns in the analysis.

```python
data['eduyrs'].hist(bins=30)
plt.title("Years of education distribution")
plt.xlabel("Years of education")
plt.ylabel("Frequency")
plt.show()
```

### Distribution of Education

The distribution of years of education is right-skewed, with a strong concentration between approximately 10 and 15 years, peaking around 12–13 years. Frequencies are noticeably lower below 10 years, indicating that lower levels of formal education are less represented in the sample.

Frequencies drop sharply beyond this central range, with relatively few observations above 20 years. A long right tail is present, reflecting variation in extended educational trajectories, with a small number of extreme values (e.g. above 30 years) that are likely outliers.

Overall, this indicates that most respondents have completed moderate levels of education, while both lower and higher levels of education are less represented in the dataset.

The variables age and years of education contain a small number of missing values (222 and 708 respectively), corresponding to approximately 0.45% and 1.43% of the dataset.

```python
plt.scatter(data['eduyrs'], data['agea'], alpha=0.3)
plt.xlabel("Years of education")
plt.ylabel("Age")
plt.title("Age vs Education")
plt.show()
```

```python
data[['agea','eduyrs']].corr()
```

### Relationship between Age and Education

The scatterplot and correlation analysis show a weak negative relationship between age and years of education, with a correlation of approximately -0.21. This indicates that older individuals tend to have slightly fewer years of education on average compared to younger individuals.

While education is typically accumulated early in life, the observed pattern may reflect generational differences in access to education. Older cohorts may have had fewer educational opportunities compared to younger cohorts, which could contribute to the slight downward trend.

However, the relationship remains weak, as substantial variation in education levels is observed across all age groups. Most observations are concentrated between approximately 10 and 20 years of education, with only a small number of higher values appearing as outliers.

Overall, the association between age and education is limited.

```python
data.groupby('cntry')['eduyrs'].mean().sort_values().tail(10).plot(kind='bar')
plt.title("Top countries by education")
plt.show()
```

### Key Insight - 1

Countries in Northern and Western Europe, such as Iceland, Ireland, Germany, the United Kingdom, the Netherlands, and the Nordic countries (e.g. Finland, Norway, Sweden, Denmark), show the highest average years of education in the dataset, with values generally ranging between approximately 13.5 and 15.5 years.

In contrast, countries such as Portugal, Serbia, and some Southern and Eastern European countries show lower average years of education, often closer to 10–12 years.

This pattern may reflect differences in national education systems, access to higher education, and socioeconomic development. Countries with strong higher education infrastructures and higher participation in tertiary education are likely to exhibit higher average years of education.

However, several limitations should be considered. The results are based on survey data and may not fully represent entire national populations. Additionally, differences in how education systems are structured across countries (e.g. duration of secondary and tertiary education) can affect comparability of “years of education” between countries. It is also important to note that these results reflect averages and do not capture variation within countries.

Overall, the results suggest a clear regional pattern, with higher average education levels concentrated in Northern and Western Europe.

```python
data['trstprl'].hist(bins=10)
plt.title("Trust in Parliament Distribution")
plt.xlabel("Trust level")
plt.ylabel("Frequency")
plt.show()
```

### Distribution of Trust in Parliament

The distribution of trust in parliament shows that responses are spread across the full range of the scale, indicating substantial variation in trust levels among respondents.

There appears to be a concentration in the mid-range values, suggesting that moderate levels of trust are relatively common. At the same time, both lower and higher trust values are also well represented, indicating that opinions are not strongly skewed in a single direction.

It should be noted that the variable includes both numeric values and labeled categories (e.g. "No trust at all", "Complete trust"), which may affect how the distribution is displayed and interpreted.

```python
trust_vars = ['trstprl', 'trstlgl', 'trstplc']

for col in trust_vars:
    data[col] = pd.to_numeric(data[col], errors='coerce')

data['trstprl'].mean()
```

### Average Trust in Parliament

The average level of trust in parliament is approximately 4.96 on a 0-10 scale, indicating slightly below moderate trust among respondents.

This aligns with the earlier distribution, where responses were spread across the full range but showed noticeable clustering in the mid-range, alongside a substantial proportion of lower trust values.

Overall, the results suggest that trust in parliament is mixed, with a tendency toward moderate to lower levels rather than high trust.

```python
data[trust_vars].mean()
```

### Average Trust Across Institutions

The average levels of trust differ across institutions. Trust in parliament is the lowest (4.96), followed by trust in the legal system (5.58), while trust in the police is the highest (6.34).

This indicates that respondents do not express uniform trust across institutions. Instead, political institutions appear to receive lower trust than legal and law-enforcement institutions.

Overall, the results suggest that institutional trust is differentiated rather than generalised, with trust in the police noticeably exceeding trust in parliament.

```python
data[trust_vars].corr()
```

### Correlation Between Trust Variables

The correlation analysis shows moderate positive relationships between the different trust variables. Trust in parliament and trust in the legal system are moderately correlated (r = 0.63), as are trust in the legal system and trust in the police (r = 0.64).

The correlation between trust in parliament and trust in the police is somewhat lower (r = 0.49), but still indicates a positive association.

These results suggest that while the variables are related and capture a common underlying concept of institutional trust, they are not identical. Each institution reflects a distinct aspect of trust, but with sufficient overlap to justify combining them into a broader trust scale.
